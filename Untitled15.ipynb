```python
from bs4 import BeautifulSoup
from selenium import webdriver

hotel_link = 'https://me.cleartrip.com/hotels/details/260110?c=&r=2,0&compId=&fr=0&ur=1&urt=featured&area=&dest_code=437812&tags=#'

driver = webdriver.Firefox()
driver.get(hotel_link)

content = driver.page_source
print(content)
```

```python
soup = BeautifulSoup(content, 'html')
print(soup)
```

```python
hotel_info = {"hotel_name" : "", "hotel_url" : "", "hotel_address" : "", "hotel_rating" : "", 
             "hotel_review_no" : "", "hotel_price" : "", 
              "hotel_info" : {"hotel_location" : "", 
                                "rooms" : "",
                                "amenities" : "", 
                                "dining" : "", 
                                "other_amenities" : "",
                                "check_in" : "", 
                                "check_out" : "", 
                                "total_rooms" : "", 
                                "floors" : ""}}
print(hotel_info)
```

```python
print(soup.prettify())
divparent = soup.find_all('div', attrs={'class':'hotelDescriptionBlock hotelDescriptionBottom'})
print(divparent)
```

```python
import re
details = ''
for info in soup.find_all('p'):
    details += info.get_text('\n')

print(details)
```
